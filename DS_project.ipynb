```python
import matplotlib.pyplot as plt
import pandas as pd

# Enter categorical values for each assignment
assignment1 = ["N", "Y","Y","N", "N","Y","N","N","Y","N","N","Y","Y", "Y","N","N","N", "Y","Y","N","N","Y","N", "Y","Y","Y","Y","N","Y","N","N","N","N","Y","Y","Y","N","Y","N","N","N","Y","Y","N","N","N","Y","Y"]
assignment2 = ["Y", "Y","Y","Y","Y","Y", "N","Y","Y","Y","Y","N","Y","N","Y","Y","Y","Y","Y","Y","Y","Y","Y","Y","Y","Y","N","N","N","Y","Y","Y","Y","Y","N","Y","Y","Y","Y","N","N","Y","Y","N","Y","Y","Y","Y"]
assignment3 = ["Y", "N","N","Y","N","N","N","Y","N","N","N","N","Y","N","Y","Y","Y","Y","Y","N","N","N","N","Y","Y","Y","N","N","N","Y","N","N","N","N","N","N","Y","Y","N","N","N","N","N","N","Y","N","N","N"]

# Create a DataFrame
df = pd.DataFrame({
    "Assignment 1": assignment1,
    "Assignment 2": assignment2,
    "Assignment 3": assignment3
})

for col in df.columns:
    plt.figure()
    category_counts = df[col].value_counts()
    bars = category_counts.plot(kind="bar")
    plt.title(f"Category Count - {col}")
    plt.xlabel("Category")
    plt.ylabel("Count")

    # Add value labels on bars
    for index, value in enumerate(category_counts):
       plt.text(index, value, str(value), ha='center', va='bottom')
    plt.show()
```
